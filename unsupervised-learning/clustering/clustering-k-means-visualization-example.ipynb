```python
# Importing required packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
df = pd.read_csv('sample_dense_cluster.csv')
df.shape
```

```python
# View top 5 rows
df.head()
```

```python
# Checking got null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Visualize clusters
sns.scatterplot(data=df, x='Col1', y='Col2');
```

Clearly there are 3 visible clusters, let's create K-Means model with 3 cluster and see how it performs.

```python
# K-Means model
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(df.values)
labels = model.predict(df.values)
```

```python
# Visualize clusters
sns.scatterplot(data=df, x='Col1', y='Col2', c=labels);
```

Clearly model was able to identify the clusters correctly, now let's plot cluster centers.

```python
# Plot centroid centers
centroids = model.cluster_centers_
plt.scatter(df.values[:,0], df.values[:,1], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='o', s=200, c='red')
plt.show()
```

The centroids are important because they are what enables KMeans to assign new, previously unseen points to the existing clusters
